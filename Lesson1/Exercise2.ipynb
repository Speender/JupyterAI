```python
# import all necessary modules
import numpy as np
import matplotlib.pyplot as plt

import matplotlib_inline.backend_inline
matplotlib_inline.backend_inline.set_matplotlib_formats('svg')
```

```python
# function (as a function)
def f(x):
  return (5*x**3)*(np.sin(x))

# derivative function
def df(x):
  return (5*x**3)*(np.cos(x))+ (15*x**2)*(np.sin(x))
```

```python
# plot the function and its derivative

# define a range for x
x = np.linspace(-10,10,2001)

# plotting
plt.plot(x,f(x), x,df(x))
plt.xlim(x[[0,-1]])
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(['y','dy'])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define your functions here (assuming f(x) and df(x) are already defined)

# Number of random starting points
n_starting_points = 6

# learning parameters
learning_rate = 0.01
training_epochs = 100

# Create an array to hold the local minima for each starting point
localmins = []

# List of colors
colors = plt.cm.jet(np.linspace(0, 1, n_starting_points))

# Gradient Descent for each starting point
for idx in range(n_starting_points):
    # random starting point
    localmin = np.random.choice(x, 1)
    
    for i in range(training_epochs):
        grad = df(localmin)
        localmin = localmin - learning_rate * grad
    
    # Store the local minimum found
    localmins.append(localmin[0])

# plot the results
plt.plot(x, f(x), x, df(x))

# Plot each found local minimum with a different color
for idx, localmin in enumerate(localmins):
    plt.plot(localmin, df(localmin), 'o', color=colors[idx], label=f'Min {idx + 1}')
    plt.plot(localmin, f(localmin), 'o', color=colors[idx])

plt.xlim(x[[0, -1]])
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(['f(x)', 'df'] + [f'f(x) min {i + 1}' for i in range(n_starting_points)])
plt.title(f'Empirical local minima: {localmins}')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define your functions here (assuming f(x) and df(x) are already defined)

# Number of random starting points
n_starting_points = 5

# learning parameters
learning_rate = 0.01
training_epochs = 100

# List of colors
colors = plt.cm.jet(np.linspace(0, 1, n_starting_points))

# Create an array to store model parameters for each starting point
all_modelparams = np.zeros((n_starting_points, training_epochs, 2))

# Gradient Descent for each starting point
for idx in range(n_starting_points):
    # random starting point
    localmin = np.random.choice(x, 1)
    
    for i in range(training_epochs):
        grad = df(localmin)
        localmin = localmin - learning_rate * grad
        all_modelparams[idx, i, 0] = localmin
        all_modelparams[idx, i, 1] = grad

# Plot the gradient over iterations for each starting point
fig, ax = plt.subplots(1, 2, figsize=(12, 4))

for idx in range(n_starting_points):
    for i in range(2):
        ax[i].plot(all_modelparams[idx, :, i], 'o-', color=colors[idx], label=f'Start {idx + 1}')
        ax[i].set_xlabel('Iteration')
        ax[i].set_title(f'Gradient Descent Paths')
    
ax[0].set_ylabel('Local minimum')
ax[1].set_ylabel('Derivative')

ax[0].legend()
ax[1].legend()

plt.show()
```

```python
# Define parameters
training_epochs = 100
learning_rates = [ 0.011, 0.211, 0.02, 0.11, 0.112]
start_val = 2.5 

# Define x range for plotting the function later
x_plot = np.linspace(-10, 10, 2001)

plt.figure(figsize=(10, 6))

for rate in learning_rates:
  # setup for this run
  localmin = np.array([start_val])
  modelparams = np.zeros((training_epochs, 1)) * np.nan

  # Gradient Descent code
  for i in range(training_epochs):
    grad = df(localmin)

    # Check for NaN/Inf gradient, which indicates explosion
    if np.isnan(grad) or np.isinf(grad):
        print(f"LR = {rate}: Exploded at iteration {i}!")
        break

    modelparams[i, 0] = localmin[0] 
    localmin = localmin - rate * grad 

    # Check if localmin itself exploded
    if np.isnan(localmin) or np.isinf(localmin):
        print(f"LR = {rate}: Exploded (localmin) at iteration {i+1}!")
        break
  else: 
      if i == training_epochs - 1 and (not np.isnan(localmin)):
           if i < training_epochs -1: 
                modelparams[i+1, 0] = localmin[0]

  # plot the local minimum history
  plt.plot(modelparams[:, 0], 'o-', label=f'LR = {rate}', markersize=3) 

plt.xlabel('Iteration')
plt.ylabel('Estimated Local Minimum (x)')
plt.title('Effect of Learning Rate on Convergence (Starting at x=2.5)')

all_vals = plt.gca().get_lines()
valid_vals = np.concatenate([line.get_ydata()[~np.isnan(line.get_ydata())] for line in all_vals])
if len(valid_vals) > 0:
    min_y = np.min(valid_vals) - 1
    max_y = np.max(valid_vals) + 1
    plt.ylim(min_y, max_y)
else:
    plt.ylim(-5, 5) 

plt.legend()
plt.grid()
plt.show()

plt.figure(figsize=(10, 4))
plt.plot(x_plot, f(x_plot), label='f(x)')
plt.plot(x_plot, df(x_plot), label='df(x)', alpha=0.7)
plt.title('Function f(x) and its Derivative df(x)')
plt.xlabel('x')
plt.ylabel('Value')
plt.ylim(-500, 500)
plt.axhline(0, color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.grid()
plt.show()
```

```python
# Define x range
x = np.linspace(-10, 10, 2001)
localmin = np.random.choice(x, 1)

# learning parameters
learning_rate = 0.011
training_epochs = 100

gradient_threshold = 0.1 

modelparams_list = []

print(f"Starting gradient descent from x = {localmin[0]:.4f}")
print(f"Using LR = {learning_rate}")
print(f"Stopping when |gradient| < {gradient_threshold}\n")

for i in range(training_epochs):
    grad = df(localmin)

    # Check for explosion first
    if np.isnan(grad) or np.isinf(grad):
        print(f"--- Exploded at iteration {i}! ---")
        break

    # Store parameters
    modelparams_list.append([localmin[0], grad[0]])

    # 3. Add the early stopping condition
    if np.abs(grad[0]) < gradient_threshold:
        print(f"--- Early stopping triggered! ---")
        print(f"Iteration: {i}")
        print(f"Gradient: {grad[0]:.6f}")
        print(f"Local Min (x): {localmin[0]:.6f}")
        break # Exit the loop

    # Update step
    localmin = localmin - learning_rate * grad

    if np.isnan(localmin) or np.isinf(localmin):
        print(f"--- Exploded (localmin) at iteration {i+1}! ---")
        modelparams_list.pop()
        break
else:
    print(f"--- Loop finished ({training_epochs} epochs) ---")
    if 'grad' in locals(): 
      print(f"Final Gradient: {grad[0]:.6f}")
      print(f"Final Local Min (x): {localmin[0]:.6f}")
    else:
      print("Loop did not run (potential issue with starting point).")


if modelparams_list:
    modelparams_early_stop = np.array(modelparams_list)

    fig, ax = plt.subplots(1, 2, figsize=(12, 4))

    # Plot 1: Local minimum (x) over iterations
    ax[0].plot(modelparams_early_stop[:, 0], 'o-')
    ax[0].set_xlabel('Iteration')
    ax[0].set_ylabel('Local minimum (x)')
    ax[0].set_title(f'Converged/Stopped in {modelparams_early_stop.shape[0]} iterations')

    # Plot 2: Derivative (gradient) over iterations
    ax[1].plot(modelparams_early_stop[:, 1], 'o-')
    ax[1].set_xlabel('Iteration')
    ax[1].set_ylabel('Derivative (gradient)')
    ax[1].axhline(gradient_threshold, color='r', linestyle='--', label='Threshold')
    ax[1].axhline(-gradient_threshold, color='r', linestyle='--')
    ax[1].set_title(f'Final gradient: {modelparams_early_stop[-1, 1]:.5f}')
    ax[1].legend()

    # Set y-limit for gradient plot based on actual values to avoid scaling issues if it explodes early
    valid_grads = modelparams_early_stop[:, 1][~np.isnan(modelparams_early_stop[:, 1])]
    if len(valid_grads) > 0:
        max_abs_grad = np.max(np.abs(valid_grads))
        ax[1].set_ylim(-max_abs_grad * 1.1, max_abs_grad * 1.1)


    plt.tight_layout()
    plt.show()
else:
    print("\nNo data to plot. Gradient descent might have failed immediately.")
```
